**Question 2:** In how many different states do the customers live in?

```python
import csv
from os.path import exists

datafile = "./sample_data/customers.csv"

if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

# Inicializamos variables
customers_list = []

with open(datafile, 'r') as f1:
    # Usamos DictReader para que sea más fácil acceder por nombre de columna
    csvreader = csv.DictReader(f1)

    for row in csvreader:
        customers_list.append(row)

# Pregunta 1: ¿Cuántos clientes hay?
total_customers = len(customers_list)
print(f"Total customers: {total_customers}")

# Pregunta 2: ¿En cuántos estados diferentes viven?
# Usamos un 'set' (conjunto) porque los sets no permiten duplicados automáticamente
states = set()
for customer in customers_list:
    states.add(customer['State'])

total_states = len(states)
print(f"Total different states: {total_states}")
```

**Question 3** What is the state with most customers?

```python
# Pregunta 3: ¿Cuál es el estado con la mayoría de los clientes?

state_counts = {}

for customer in customers_list:
    state = customer['State']
    if state in state_counts:
        state_counts[state] += 1
    else:
        state_counts[state] = 1

# Buscamos el estado con el valor más alto
top_state = max(state_counts, key=state_counts.get)
count_top_state = state_counts[top_state]

print(f"El estado con más clientes es: {top_state} ({count_top_state} clientes)")
```

**Question 4** What is the state with the least customers?

```python
# Pregunta 4: ¿Cuál es el estado con la menor cantidad de clientes?

# Usamos min() en lugar de max()
bottom_state = min(state_counts, key=state_counts.get)
count_bottom_state = state_counts[bottom_state]

print(f"El estado con menos clientes es: {bottom_state} ({count_bottom_state} cliente/s)")
```

**Question 5:** What is the most common last name?

```python
# Pregunta 5: ¿Cuál es el apellido más común?

last_name_counts = {}

for customer in customers_list:
    last_name = customer['LastName']
    if last_name in last_name_counts:
        last_name_counts[last_name] += 1
    else:
        last_name_counts[last_name] = 1

# Buscamos el apellido con el valor más alto
common_last_name = max(last_name_counts, key=last_name_counts.get)
count_last_name = last_name_counts[common_last_name]

print(f"El apellido más común es: {common_last_name} ({count_last_name} personas)")
```

```python
import csv
from collections import Counter

orders_file = "./sample_data/orders.csv"
orders_list = []

# Cargamos los datos
with open(orders_file, 'r') as f2:
    csvreader = csv.DictReader(f2)
    for row in csvreader:
        orders_list.append(row)

# --- Pregunta 1: Órdenes únicas ---
unique_orders = set(row['OrderID'] for row in orders_list)
total_orders = len(unique_orders)

# --- Pregunta 2: Promedio de elementos por orden ---
items_per_order = Counter(row['OrderID'] for row in orders_list)
avg_items = sum(items_per_order.values()) / len(items_per_order)

# --- Pregunta 3: Mayor número de artículos por pedido ---
max_items = max(items_per_order.values())

# --- Pregunta 4: Pedidos en Octubre 2021 ---
# Filtramos por las órdenes únicas que tienen "2021-10" en la fecha
oct_2021_orders = set(row['OrderID'] for row in orders_list if "2021-10" in row['Date'])
total_oct_2021 = len(oct_2021_orders)

# --- Pregunta 5: Cliente que más gastó en 2021 ---
customer_spending = {}
processed_orders = set()
for row in orders_list:
    if "2021" in row['Date']:
        oid = row['OrderID']
        if oid not in processed_orders:
            cid = row['CustomerID']
            total = float(row['OrderTotal'])
            customer_spending[cid] = customer_spending.get(cid, 0) + total
            processed_orders.add(oid)

top_customer = max(customer_spending, key=customer_spending.get)
max_spend = customer_spending[top_customer]

# --- Pregunta 6: Mejor mes para las ventas (histórico) ---
month_sales = {}
processed_orders_months = set()
for row in orders_list:
    oid = row['OrderID']
    if oid not in processed_orders_months:
        # Truco: tomamos solo los primeros 7 caracteres (ej: "2021-01")
        # Esto evita errores con las horas y milisegundos
        month = row['Date'][:7]
        if len(month) == 7 and month[4] == '-': # Validamos que sea formato YYYY-MM
            total = float(row['OrderTotal'])
            month_sales[month] = month_sales.get(month, 0) + total
            processed_orders_months.add(oid)

best_month = max(month_sales, key=month_sales.get)

# --- RESULTADOS FINALES ---
print(f"1. Órdenes únicas: {total_orders}")
print(f"2. Promedio de elementos: {round(avg_items, 2)}")
print(f"3. Máximo de artículos: {max_items}")
print(f"4. Pedidos Octubre 2021: {total_oct_2021}")
print(f"5. Cliente que más gastó en 2021: {top_customer}")
print(f"6. Mejor mes histórico: {best_month}")
```
